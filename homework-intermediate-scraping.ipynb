### 1) Grab the HTML for the webpage linked above

Use `requests` to get the HTML, assigning it to a variable

```python
data = requests.get("https://en.wikipedia.org/wiki/List_of_tallest_buildings_in_Brooklyn")
textData = data.text
#textData
```

### 2) Convert the HTML into a `BeautifulSoup` object

```python
soupedData = bs(textData)
print(soupedData)
```
